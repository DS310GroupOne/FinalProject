```python
import pandas as pd
```

```python
df = pd.read_csv('Stats.csv')
```

```python
df = df.drop(columns = 'Unnamed: 0')
```

```python
df['Confirmed_Change_30d_Rolling_Adjusted'] = df.groupby('CountryName')['Confirmed_Change'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
df['Deaths_Change_30d_Rolling_Adjusted'] = df.groupby('CountryName')['Deaths_Change'].transform(lambda x: x.rolling(window=30, min_periods=1).mean())
df.head(30)
```

```python
data = df
```

```python
data['Confirmed_Growth_Rate'] = data.groupby('CountryName')['Confirmed_Change_30d_Rolling_Adjusted'].pct_change().replace([float('inf'), -float('inf')], None)
data['Deaths_Growth_Rate'] = data.groupby('CountryName')['Deaths_Change_30d_Rolling_Adjusted'].pct_change().replace([float('inf'), -float('inf')], None)

data['Confirmed_Growth_Rate'].fillna(0, inplace=True)
data['Deaths_Growth_Rate'].fillna(0, inplace=True)

data[['CountryName', 'Confirmed_Change_30d_Rolling_Adjusted', 'Confirmed_Growth_Rate', 'Deaths_Change_30d_Rolling_Adjusted', 'Deaths_Growth_Rate']].head()
```

```python
data
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

data['Confirmed_Growth_Binary'] = (data['Confirmed_Change_30d_Rolling_Adjusted'] > 3).astype(int)
data['Deaths_Growth_Binary'] = (data['Deaths_Change_30d_Rolling_Adjusted'] > 1).astype(int)

predictors = ['C1_School_closing', 'C2_Workplace_closing',
       'C3_Cancel_public_events', 'C4_Restrictions_on_gatherings',
       'C5_Close_public_transport', 'C6_Stay_at_home_requirements',
       'C7_Restrictions_on_internal_movement',
       'C8_International_travel_controls', 'E1_Income_support',
       'E2_Debt_contract_relief', 'H1_Public_information_campaigns',
       'H2_Testing_policy', 'H3_Contact_tracing', 'H6_Facial_Coverings', 'GovernmentResponseIndex',
            'ContainmentHealthIndex',
       'H7_Vaccination_policy']

data_clean = data.dropna(subset=predictors + ['Confirmed_Growth_Binary', 'Deaths_Growth_Binary'])

logreg_cases = LogisticRegression(max_iter=1000)
logreg_cases.fit(data_clean[predictors], data_clean['Confirmed_Growth_Binary'])
case_coefficients = pd.DataFrame({'Predictor': predictors, 'Coefficient': logreg_cases.coef_[0], 'Odds Ratio': np.exp(logreg_cases.coef_[0])})

logreg_deaths = LogisticRegression(max_iter=1000)
logreg_deaths.fit(data_clean[predictors], data_clean['Deaths_Growth_Binary'])
death_coefficients = pd.DataFrame({'Predictor': predictors, 'Coefficient': logreg_deaths.coef_[0], 'Odds Ratio': np.exp(logreg_deaths.coef_[0])})

case_coefficients, death_coefficients
```

```python
case_coefficients.nsmallest(6, ['Coefficient']).reset_index(drop= True)[1:6]
```
